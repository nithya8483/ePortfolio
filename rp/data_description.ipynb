```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Gold_price_gbp.xlsx")
```

```python
df.head()
```

```python
# Setting Date column as an index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index("Date" , inplace = True)
```

```python
plt.figure(figsize=(15, 6))
df['GBP'].plot()
plt.ylabel("Price in GBP")
plt.xlabel("Date")
plt.title("Gold Price")
plt.tight_layout()
plt.show()
```
